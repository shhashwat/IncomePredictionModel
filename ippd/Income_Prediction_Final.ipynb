```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df1= df.copy()
df1= df1.apply(LabelEncoder().fit_transform)
df1.head()
```

```python
ss= StandardScaler().fit(df1.drop('income', axis=1))
```

```python
X= ss.transform(df1.drop('income', axis=1))
y= df['income']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression()

model = lr.fit(X_train, y_train)
prediction = model.predict(X_test)

print("Acc on training data: {:,.3f}".format(lr.score(X_train, y_train)*100))
print("Acc on test data: {:,.3f}".format(lr.score(X_test, y_test)*100))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
}

grid_search_lr = GridSearchCV(LogisticRegression(), param_grid, cv=5, scoring='accuracy')

grid_search_lr.fit(X_train, y_train)
print("Best Logistic Regression Hyperparameters:", grid_search_lr.best_params_)
best_lr_model = grid_search_lr.best_estimator_
y_pred_lr = best_lr_model.predict(X_test)
test_accuracy_lr = accuracy_score(y_test, y_pred_lr)
print("Logistic Regression Test Accuracy with Best Hyperparameters:", test_accuracy_lr)

```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svm=SVC()

model=svm.fit(X_train,y_train)
prediction = model.predict(X_test)
accuracy_svm=svm.score(X_train, y_train)
print("Acc on training data: {:,.3f}".format(svm.score(X_train, y_train)*100))
print("Acc on test data: {:,.3f}".format(svm.score(X_test, y_test)*100))
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = 3
knn_classifier = KNeighborsClassifier(n_neighbors=k)
knn_classifier.fit(X_train, y_train)
prediction = knn_classifier.predict(X_test)
accuracy = accuracy_score(y_test, prediction)
print("Accuracy:", accuracy)
print("Acc on test data: {:,.3f}".format(knn_classifier.score(X_test, y_test)*100))
```

```python
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
}
grid_search_knn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search_knn.fit(X_train, y_train)
print("Best K-Nearest Neighbors Hyperparameters:", grid_search_knn.best_params_)
best_knn_model = grid_search_knn.best_estimator_
y_pred_knn = best_knn_model.predict(X_test)
test_accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("K-Nearest Neighbors Test Accuracy with Best Hyperparameters:", test_accuracy_knn)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=200, random_state=42)
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Acc on test data: {:,.3f}".format(rf_classifier.score(X_test, y_test)*100))
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 4],
}

grid_search_rf = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search_rf.fit(X_train, y_train)
print("Best Random Forest Hyperparameters:", grid_search_rf.best_params_)
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)
test_accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Test Accuracy with Best Hyperparameters:", test_accuracy_rf)
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
xgb_classifier = XGBClassifier(booster='gbtree', n_estimators=100, max_depth=3, learning_rate=0.1, random_state=42)

xgb_classifier.fit(X_train, y_train)

y_pred = xgb_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Acc on test data: {:,.3f}".format(xgb_classifier.score(X_test, y_test)*100))



```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.01, 0.001]
}
xgb_classifier = XGBClassifier(booster='gbtree', random_state=42)
grid_search = GridSearchCV(estimator=xgb_classifier, param_grid=param_grid, scoring='accuracy', cv=5)
grid_search.fit(X_train, y_train)
best_xgb_model = grid_search.best_estimator_
y_pred = best_xgb_model.predict(X_test)
accuracy_xgb_test = accuracy_score(y_test, y_pred)
print("Best XGBoost Model Accuracy:", accuracy_xgb_test)
```
